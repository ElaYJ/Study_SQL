문제 6.

위도, 경도 정보를 이용하여 미왕빌딩에서 1킬로 이내에 위치한 주유소 정보를 검색하세요.

- 주유소 아이디, 주유소 브랜드명, 주유소 매장명, 주소, 미왕빌딩으로 부터의 거리 (km)

제출 6.
- 쿼리, 결과

MySQL 위도 경도를 활용해서 거리를 계산하는 구문을 작성

- 출처: https://yusang.tistory.com/48

````python
sql = '''
SELECT s.id, b.name, s.address, 
(6371*acos(cos(radians(s.lat))*cos(radians(37.4955699))*cos(radians(127.0293726)-radians(s.lng))+sin(radians(s.lat))*sin(radians(37.4955699))))
AS distance FROM gas_station s, gas_brand b WHERE s.brand = b.id HAVING distance<=1
'''
````

````python
'''
위도와 경도로 거리를 구하는 방법에는 여러 가지 방법이 있지만, 가장 널리 사용되는 방법 중 하나는 Haversine 공식을 이용하는 것입니다.
이 공식은 지구상의 두 지점 간의 직선 거리를 계산하는 데 사용됩니다.

Haversine 공식은 다음과 같이 표현됩니다:

```
a = sin²(Δφ/2) + cos φ1 * cos φ2 * sin²(Δλ/2)
c = 2 * atan2( √a, √(1-a) )
d = R * c
```
여기서,
- φ1, φ2: 두 지점의 위도
- λ1, λ2: 두 지점의 경도
- Δφ = φ2 - φ1
- Δλ = λ2 - λ1
- R: 지구의 반지름 (보통 평균 지구 반지름인 6371km를 사용)

이 공식을 사용하여 두 지점 사이의 거리를 계산할 수 있습니다.
이를 위해 일반적으로 프로그래밍 언어의 수학 함수를 사용하여 삼각함수 및 역삼각함수를 계산합니다.

예를 들어 Python에서는 다음과 같이 구현할 수 있습니다.
이 코드는 두 지점 간의 거리를 킬로미터 단위로 계산합니다.
'''

from math import radians, sin, cos, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # 지구의 반지름 (km)

    # 위도와 경도를 라디안으로 변환
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    # 위도와 경도의 차이를 계산
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    # Haversine 공식을 이용하여 거리 계산
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c

    return distance

# 예시: 두 지점 간의 거리 계산 (서울과 뉴욕의 경우)
seoul_lat, seoul_lon = 37.5665, 126.9780  # 서울 위도, 경도
new_york_lat, new_york_lon = 40.7128, -74.0060  # 뉴욕 위도, 경도

distance = haversine(seoul_lat, seoul_lon, new_york_lat, new_york_lon)
print("두 지점 간의 거리 (km):", distance)
````

````python
# 미왕빌딩 37.4955366 127.0293521
sql = '''
SELECT s.id, b.name brand, s.name, address, gasoline, self, 24_hours, convenience_store, 
(6371*acos(cos(radians(37.4955366))*cos(radians(s.lat))*cos(radians(s.lng)-radians(127.0293521))+sin(radians(37.4955366))*sin(radians(s.lat)))) distance 
FROM gas_brand b, gas_station s 
WHERE b.id=s.brand AND (self=1 AND 24_hours=1 AND convenience_store=1) 
ORDER BY distance LIMIT 10;
'''
query = """
SELECT s.id, b.name brand, s.name, s.address, s.gasoline, s.self, s.24_hours, s.convenience_store, s.distance_km 
FROM gas_brand b, (SELECT id, brand, name, address, gasoline, self, 24_hours, convenience_store, 
(6371*acos(cos(radians(37.4955366))*cos(radians(lat))*cos(radians(lng)-radians(127.0293521))+sin(radians(37.4955366))*sin(radians(lat)))) distance_km 
FROM gas_station WHERE self=1 AND 24_hours=1 AND convenience_store=1 ORDER BY distance_km LIMIT 10) s
WHERE b.id=s.brand ORDER BY gasoline;
"""
cursor.execute(query)
df = pd.DataFrame(cursor.fetchall())
df.columns = cursor.column_names
df
````
